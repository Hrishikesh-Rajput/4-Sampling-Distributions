```python
import random
import matplotlib.pyplot as plt
```

```python
# Assuming you have a database of shopper IDs
shopper_ids = list(range(1, 10001))
```

```python
# Perform simple random sampling
sample_size = 300
```

```python
random_sample = random.sample(shopper_ids, sample_size)
```

```python
# Simulate satisfaction scores (for illustration)
satisfaction_scores = [random.randint(1, 10) for _ in range(sample_size)]
```

```python
# Plot a histogram of satisfaction scores
plt.figure(figsize=(10, 6))
plt.hist(satisfaction_scores, bins=range(1, 12), align='left', alpha=0.7, edgecolor='black')
plt.title('Distribution of Satisfaction Scores in Randomly Sampled Shoppers')
plt.xlabel('Satisfaction Score')
plt.ylabel('Frequency')
plt.xticks(range(1, 11))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
